```python
# Load in SQL, Pandas and SQLite

%load_ext sql
%sql
import pandas as pd
import sqlite3
%sql sqlite:///CourseData.db 
conn = sqlite3.connect('CourseData.db')
```

```python


%%sql

DROP TABLE IF EXISTS LOCATION;
CREATE TABLE LOCATION (
    LocationID INTEGER PRIMARY KEY,
    Location TEXT,
    Term TEXT NOT NULL,
    Section TEXT
);
```

```python


%%sql

DROP TABLE IF EXISTS INSTRUCTOR;
CREATE TABLE INSTRUCTOR (
    InstructorID INTEGER PRIMARY KEY,
    InstructorName TEXT NOT NULL
);
```

```python


%%sql

DROP TABLE IF EXISTS PROGRAM;
CREATE TABLE PROGRAM (
    ProgramID INTEGER PRIMARY KEY,
    ProgramCode TEXT NOT NULL,
    ProgramName TEXT NOT NULL
);
```

```python

%%sql

DROP TABLE IF EXISTS COURSE;
CREATE TABLE COURSE (
    CatalogID TEXT PRIMARY KEY,
    ProgramID
    CourseTitle TEXT NOT NULL,
    Description TEXT,
    Attributes TEXT,
    Coreqs TEXT,
    Prereqs TEXT,
    Fees TEXT,
    Credits INTEGER NOT NULL,
    FOREIGN KEY (ProgramID) REFERENCES PROGRAM (ProgramID)
);
```

```python


%%sql

DROP TABLE IF EXISTS COURSE_MEETING;
CREATE TABLE COURSE_MEETING (
    CourseMeetingID INTEGER PRIMARY KEY,
    CatalogID TEXT NOT NULL,
    Location TEXT,
    StartDateTime TEXT,
    EndDateTime TEXT,
    FOREIGN KEY (Location) REFERENCES LOCATION (Location)
);
```

```python

%%sql

DROP TABLE IF EXISTS COURSE_OFFERING;
CREATE TABLE COURSE_OFFERING (
    CourseOfferingID INTEGER PRIMARY KEY,
    CatalogID TEXT NOT NULL,
    CourseTitle TEXT NOT NULL,
    InstructorName TEXT NOT NULL,
    Section TEXT, 
    Meetings TEXT NOT NULL,
    Timecodes TEXT NOT NULL,
    Term TEXT NOT NULL,
    StartDay TEXT NOT NULL,
    End TEXT NOT NULL,    
    Cap INTEGER NOT NULL,
    Act INTEGER NOT NULL,
    Rem INTEGER NOT NULL,
    FOREIGN KEY (CatalogID) REFERENCES COURSE_MEETING (CatalogID)
    FOREIGN KEY (InstructorName) REFERENCES INSTRUCTOR 
    (InstructorName)
    FOREIGN KEY (CourseTitle) REFERENCES COURSE (CourseTitle)
);


```

```python
%%sql 
DELETE FROM LOCATION;

INSERT INTO LOCATION (LocationID, Location, Term, Section)
SELECT DISTINCT Location, Term, Section
FROM CourseData
```

```python
%%sql
DELETE FROM INSTRUCTOR;

INSERT INTO INSTRUCTOR (InstructorID, InstructorName)
SELECT DISTINCT primary_instructor
FROM CourseData
```

```python
%%sql
DELETE FROM PROGRAM;

INSERT INTO PROGRAM (ProgramID, ProgramCode, ProgramName)
SELECT DISTINCT program_code, program_name
FROM CourseData
```

```python
%%sql
DELETE FROM COURSE;

INSERT INTO COURSE (CatalogID, ProgramID, CourseTitle, Description, Attributes, Coreqs, Prereqs, Fees, Credits)
SELECT DISTINCT catalog_id, PROGRAM.ProgramID, course_title, description, attributes, coreqs, prereqs, fees, credits
FROM CourseData
    JOIN PROGRAM ON (PROGRAM.ProgramCode = CourseData.program_code)
```

```python
%%sql
DELETE FROM COURSE_MEETING

INSERT INTO COURSE_MEETING (CourseMeetingID, CatalogID, Location, StartDateTime, EndDateTime)
SELECT DISTINCT catalog_id, location, start, end
FROM CourseData
    JOIN LOCATION ON (LOCATION.Location = CourseData.location)
```

```python
%%sql
DELETE FROM COURSE_OFFERING

INSERT INTO COURSE_OFFERING (CourseOfferingID, CatalogID, CourseTitle, InstructorName, Section, Meetings, TimeCodes, Term, StartDay, End, Cap, Act, Rem)
SELECT DISTINCT catalog_id, course_title, primary_instructor, section, meetings, timecodes, term, start, end, cap, act, rem
FROM CourseData
    JOIN COURSE_MEETING ON (COURSE_MEETING.CatalogID = CourseData.catalog_id)
    JOIN INSTRUCTOR ON (INSTRUCTOR.InstructorName = CourseData.primary_instructor)
    JOIN COURSE ON (COURSE.CourseTitle = CourseData.course_title)
```
